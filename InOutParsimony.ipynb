```python
from InOutParsimony import InOutParsimony, SolutionTreeToPrint, SolutionTreeToWrite
from sowing.repr.newick import parse, write
import json
```

```python
#Smaller Example

#INPUT

#Synteny tree
syntenyTree = parse("(((1,2),(3,4)),(((5,6),7),8));")

#Content dictionnary
x_tilde = {"1": {"a","b"},
           "2": {"a","c"},
           "3": {"b"},
           "4": {"a"},
           "5": {"c"},
           "6": {"c"},
           "7": {"a","c","d"},
           "8": {"d"}}

#Costs of events
cost_loss = 1
cost_gain = 1
```

```python
cost, x, positionGainLoss = InOutParsimony(syntenyTree, x_tilde, cost_loss, cost_gain)
```

```python
print(SolutionTreeToPrint(syntenyTree, x, positionGainLoss))
```

```python
print(SolutionTreeToWrite(syntenyTree, x, positionGainLoss))
```
